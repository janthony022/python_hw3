```python
# 1) Build a Shopping Cart
# You can use either lists or dictionaries. The program should have the following capabilities:

# 1) Takes in input
# 2) Stores user input into a dictionary or list
# 3) The User can add or delete items
# 4) The User can see current shopping list
# 5) The program Loops until user 'quits'
# 6) Upon quiting the program, print out all items in the user's list
# from IPython.display import clear_output

shopping_list = ["beef", "chicken", "fish", "pork", "apple", "banana"]

def main():
    while True:
        print(''' ### SHOPPING CART ###
        1. View Original List
        2. Add Item
        3. Delete Item
        4. check items on the list
        ''')
        
        def costco():
            print()
            for i in shopping_list:
                print(i)

        def additem():
            item = input("Add to list: ")
            shopping_list.append(item)
            print(item + " added")

        def removeitem():
            item = input("Which item from the list: ")
            shopping_list.remove(item)
            print(item + " has been removed.")
        
        def check_list():
            item = input("what would you like to check on the shopping list: ")
            if item in shopping_list:
                print("Yes, " + (item) + " is on the shopping list")
            else:
                print("No, " + (item) + " is not on the shopping list")
        
        def quit():
            for i in shopping_list:
                print(i)

        selection = input("input#: ")

        if selection == "1":
            costco()
        elif selection == "2":
            additem()
        elif selection == "3":
            removeitem()
        elif selection == "4":
            check_list()
        elif selection == "quit":
            print(shopping_list)
            break

main()

# 2) Create a Module in VS Code and Import It into jupyter notebook
# Module should have the following capabilities:

# 1) Has a function to calculate the square footage of a house
# Reminder of Formula: Length X Width == Area

length = int(input("input length: "))
width = int(input("input width: "))

def sqft(length, width):
    area = length*width
    return area

print("The squarefootage of the home is", sqft(length, width), "sqft.")

# 2) Has a function to calculate the circumference of a circle
radius = float(input("Radius: "))

def circle():
    circumference = 2 * 3.14 * radius
    return circumference

print("Circumference of a circle:", circle())

# Program in Jupyter Notebook should take in user input and use imported functions to calculate a circle's circumference or a houses square footage
    
```
